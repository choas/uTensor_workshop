# 01 - XOR Modell mit Keras

```python
import sys
import numpy as np
import tensorflow as tf
import keras.backend as K
from keras.models import Sequential
from keras.layers import Dense
from keras.activations import sigmoid
from keras.activations import relu
from keras.activations import tanh
from keras.losses import MSE
from keras.losses import mean_squared_error
#from keras.losses import binary_crossentropy
from keras.optimizers import SGD, Adam
from keras.metrics import binary_accuracy

from tensorflow.python.framework.graph_util import convert_variables_to_constants

import keras
print keras.__name__, keras.__version__
print tf.__name__, tf.__version__
```

Keras sollte die Version 2.2.2 und TensorFlow 1.12.0 haben.

#### Tensorflow und Keras Session anlegen

```python
sess=tf.Session()
K.set_session(sess)
```

#### Trainings- und Testdaten anlegen

```python
training_data = np.array([[0,0], [0,1], [1,0], [1,1]])
target_data   = np.array([  [0],   [1],   [1],   [0]])
```

#### XOR modell anlegen

```python
model = Sequential()
model.add(Dense(4, input_dim=2, activation='sigmoid', name='inputlayer'))
model.add(Dense(1, activation='sigmoid', name='outputlayer'))
model.compile(loss='mean_squared_error', optimizer=SGD(lr=1), metrics=['accuracy'])
```

#### TensorBoard Ausgabe konfigurieren

```python
from keras.callbacks import TensorBoard
from tensorflow.contrib.tensorboard.plugins import projector
tensorboard = TensorBoard(log_dir='./logs/xor_logs/keras', histogram_freq=0,
                          write_graph=True, write_images=False)
```

#### Training starten

```python
epochs = 2000
model.fit(training_data, target_data, epochs=epochs, callbacks=[tensorboard])
```

#### Ergebnisse

```python
print "loss:", model.evaluate(x=training_data, y=target_data, verbose=0)
print ""
print model.predict(training_data)
print ""
print model.predict(training_data).round()
```

## freeze Modell und als TensorFlow Datei speichern

```python
freeze_var_names = list(set(v.op.name for v in tf.global_variables()))
print freeze_var_names
output_names = [model.outputs[0].name.split(":")[0]]

from tensorflow.python.framework.graph_util import remove_training_nodes
sub_graph_def = remove_training_nodes(sess.graph_def)

frozen_graph = convert_variables_to_constants(sess,
                                              sub_graph_def,
                                              output_names,
                                              freeze_var_names)

graph_path = tf.train.write_graph(frozen_graph, "models", "xor_keras.pb", as_text=False)
print output_names
print('%s written' % graph_path)
```
